## Ejemplo nº1. La función logística.

El motivo de que Python sea tan utilizado en el campo del Machine Learning es
gracias a la gran cantidad de librerías disponibles. Sin embargo, para un 
adecuado aprendizaje, es interesante verificar esas librerías y sus 
manuales. En el siguiente ejercicio, se verificará que la función
logística (o sigmoide) ofrecida por la librería Scipy (scipy.special.expit()) es correcta.

La función logística o sigmoide $ \phi(x) = \frac{1}{1 + \exp(-x)} $ es normalmente utilizada para mapear números reales en probabilidades en el rango de (0, 1). Esta función se utilizará en futuras etapas.


```
# Numpy: Librería muy util que implementa muchas funcionalidades para el cálculo 
# numérico con vectores, matrices, etc.
import numpy as np
import scipy as sc

# Librería para la visualización de datos.
import matplotlib.pyplot as plt

####

sigmoid = lambda x: 1 / (1 + np.exp(-x))

# La siguiente función generará un array de 100 valores de -10 a 10.
value = np.linspace(-10, 10, 100)

plt.plot(value, sc.special.expit(value), "b--")
plt.plot(value, sigmoid(value), "r*")

plt.show()
```

```
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

###

def random_normal_var(mean, variance, amount):
  return sc.stats.norm(mean, np.sqrt(variance)).rvs(amount)

result = random_normal_var(5,4,10000)

plt.xlabel("Values")
plt.hist(result, bins=250)
plt.show()
```

```
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import scipy as sc

iris = load_iris()

X = iris.data
Y = iris.target

# print(iris.DESCR)

def statistic_table():
  result = sc.stats.describe(X)
  for column in range(len(X[0])):
    # Se agrega el 'end = " ", flush=True' para obtener todos los resultados en una sola línea.
    # Es posible usar otros métodos para obtener múltiples variables en una sola línea,
    # como los que se verán en futuras etapas.
    print("Col", column+1,end=" ", flush=True)
    print("MinVal:",round_val(result.minmax[0][column]),end=" ", flush=True)
    print("MaxVal:",round_val(result.minmax[1][column]),end=" ", flush=True)
    print("Mean:",round_val(result.mean[column]),end=" ", flush=True)
    print("Variance:",round_val(result.variance[column]))
  
def round_val(number):
  return sc.around(number,decimals=3)

print("Apartado 1")
print("")
statistic_table()
print("")

def petal_length_hist():
  plt.xlabel("Petal Length histogram")
  plt.hist(X[Y==0,2], bins=10, alpha=0.5, label="Plant 0")
  plt.hist(X[Y==1,2], bins=10, alpha=0.5, label="Plant 1")
  plt.hist(X[Y==2,2], bins=10, alpha=0.5, label="Plant 2")
  plt.show()
    
print("Apartado 2")    
petal_length_hist()
print("")


def scatter_plot():
  plt.xlabel("Sepal Length")
  plt.ylabel("Petal Width")
  plt.scatter(X[Y==0,0], X[Y==0,3])
  plt.scatter(X[Y==1,0], X[Y==1,3])
  plt.scatter(X[Y==2,0], X[Y==2,3])

print("Apartado 3")
scatter_plot()
```
